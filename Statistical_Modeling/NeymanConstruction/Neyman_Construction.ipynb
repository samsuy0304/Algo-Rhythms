```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate a coin flip experiment
def coin_flip_experiment(true_probability, num_flips=100):
    outcomes = np.random.rand(num_flips) < true_probability
    return outcomes.sum()

# Parameters
true_parameter_values = np.linspace(0, 1, 100)  # True probability values from 0 to 1
num_experiments_per_parameter = 1000  # Number of experiments for each true parameter value

# Simulate experiments and collect results
measured_values = []
for true_prob in true_parameter_values:
    measurements = [coin_flip_experiment(true_prob) for _ in range(num_experiments_per_parameter)]
    measured_values.extend([(true_prob, measurement) for measurement in measurements])

# Convert the measured values to a NumPy array for plotting
measured_values = np.array(measured_values)

# Plotting Neyman Construction
plt.figure(figsize=(10, 6))
plt.scatter(measured_values[:, 0], measured_values[:, 1], alpha=0.5, color='blue')
plt.xlabel('True Probability')
plt.ylabel('Measured Value')
plt.title('Neyman Construction: True vs. Measured Values')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Function to simulate a coin flip experiment
def coin_flip_experiment(true_probability, num_flips=100):
    outcomes = np.random.rand(num_flips) < true_probability
    return outcomes.sum()

# Parameters
true_parameter_values = np.linspace(0, 1, 100)  # True probability values from 0 to 1
num_experiments_per_parameter = 1000  # Number of experiments for each true parameter value

# Simulate experiments and collect results
measured_values = []
for true_prob in true_parameter_values:
    measurements = [coin_flip_experiment(true_prob) for _ in range(num_experiments_per_parameter)]
    measured_values.extend([(true_prob, measurement) for measurement in measurements])

# Convert the measured values to a NumPy array for plotting
measured_values = np.array(measured_values)

# Bonus: Pretend you measured a specific value
measured_value_to_estimate = 40

# Extract the true parameter values corresponding to the measured value
posterior_samples = measured_values[measured_values[:, 1] == measured_value_to_estimate, 0]

# Plotting Neyman Construction
plt.figure(figsize=(10, 6))
plt.scatter(measured_values[:, 0], measured_values[:, 1], alpha=0.5, color='blue')
plt.axhline(y=measured_value_to_estimate, color='red', linestyle='--', label='Measured Value')
plt.xlabel('True Probability')
plt.ylabel('Measured Value')
plt.title('Neyman Construction: True vs. Measured Values')
plt.legend()
plt.show()

# Plotting Posterior Distribution
plt.figure(figsize=(8, 5))
plt.hist(posterior_samples, bins=30, density=True, alpha=0.7, color='green', label='Posterior Distribution')
plt.xlabel('True Probability')
plt.ylabel('Density')
plt.title(f'Posterior Distribution for Measured Value = {measured_value_to_estimate}')
plt.legend()
plt.show()
```
