```python
import numpy as np
import matplotlib.pyplot as plt

lam = 2.0  # rate parameter for exponential distribution
```

```python
# Define parameters
N = 10  # number of samples per experiment
M = 1000  # number of experiments


# Generate M experiments, each with N samples from exponential distribution
samples = np.random.exponential(scale=1/lam, size=(M, N))

# Compute sample means for each experiment
sample_means = np.mean(samples, axis=1)

# Plot histogram of sample means
plt.hist(sample_means, bins=50, density=True)

# Plot theoretical Gaussian distribution
mu = 1/lam  # true population mean
sigma = mu / np.sqrt(N)  # true population standard deviation
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 1000)
plt.plot(x, 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2/(2*sigma**2)))

plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.title(f'Sampling Distribution of Sample Means (N={N}, M={M}, $\lambda$={lam})')

plt.show()
```
