```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Define the function to be minimized
def quadratic_function(x):
    return x**2 + 3*x + 5  # Example quadratic function with a minimum

# Visualization function for plotting the function and minimum
def plot_function_and_minimum(func, minimum_location):
    x_values = np.linspace(-10, 5, 100)
    y_values = func(x_values)

    plt.plot(x_values, y_values, label='Function to Minimize')
    plt.scatter(minimum_location, func(minimum_location), color='red', label='Minimum Location')
    plt.title('Function and Minimum')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

# Find the minimum using scipy.optimize.minimize
initial_guess = 0  # Initial guess for the minimum
result = minimize(quadratic_function, initial_guess)

# Display the result
print("Minimum found at x =", result.x)
print("Minimum function value =", result.fun)

# Visualize the function and the location of the minimum
plot_function_and_minimum(quadratic_function, result.x)
```
