```python
import numpy as np
from sklearn.neighbors import KernelDensity
```

```python
# Simulate dataset sampled from the sine function
np.random.seed(42)
n_samples = 100
X = np.sort(2 * np.pi * np.random.rand(n_samples))
y = np.sin(X) + 0.2 * np.random.randn(n_samples)
```

```python
#define the Gaussian Kernel density estimator function.
def gaussian_kernel_density_estimation(X, bandwidth):
    kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth).fit(X[:, np.newaxis])
    x_grid = np.linspace(0, 2*np.pi, 1000)
    log_density = kde.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_density), x_grid
```

```python
bandwidth = 0.2
density, x_grid = gaussian_kernel_density_estimation(X, bandwidth)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.hist(X, bins=25, density=True, alpha=0.5, label="Simulated Data")
ax.plot(x_grid, density, label="Density Estimate")
ax.set_title("Gaussian Kernel Density Estimation")
ax.set_xlabel("x")
ax.set_ylabel("Density")
ax.legend()
plt.show()
```
